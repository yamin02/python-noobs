```python
#find minmum and max after a start e.g. point1 and point 2
# for bounds also use second graph

from scipy import *
from scipy import optimize
import matplotlib.pyplot as plt

def f(x):
    return sin(x) -  (2 * exp(-(x - 0.4)**2))

x = arange(-7, 7, 0.1)
y = f(x)

point1 = -1.0 
point2 = 3.0 

min1 = optimize.fmin(f, point1)
min2 = optimize.fmin(f, point2)

minima = optimize.fminbound(f, point1, point2)

fig, ax = plt.subplots(2,1,figsize=(12, 8)) 
ax[0].plot(x, y, label = '$\sin(x)-2e^{-(x-0.4)^2}$')  # plotting curve

ax[0].plot(min1, f(min1), 's', label='min 1', marker='v', color='r')    # plotting min1 
ax[0].plot(point1, f(point1), 's', label='start 1', marker='o', color='r')     # plotting starting point for min1

ax[0].plot(min2, f(min2) , 's', label='min 2', marker='v', color='g')    # plotting min2 
ax[0].plot(point2, f(point2), 's', label='start 2', marker='o', color='g')     # plotting starting point for min2

ax[0].set_title('Start(given) minimum')
ax[0].grid()
ax[0].legend(loc='best')

ax[1].plot(x,y)
ax[1].plot(point1,f(point1),color ='red', marker = 'v', label = "start")
ax[1].plot(point2,f(point2),marker='v', label = "End")
ax[1].plot(minima,f(minima),marker="o", label = "THE MINIMUM")
ax[1].legend(loc='best')
ax[1].set_title('Bounded Minimum')
#so that two graph title or other things dont overlap each other
fig.tight_layout()
```
